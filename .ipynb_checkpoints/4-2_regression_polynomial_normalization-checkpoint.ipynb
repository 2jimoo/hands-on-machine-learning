# 비선형 데이터를 선형모델로 예측하기
### 다항회귀
- n승 데이터 필드를 추가하여 선형 모델 학습하는 방법
- PolynomialFeatures

```python
import numpy as np

m= 100
X= 6*np.random.rand(m,1) - 3
y= 0.5* X**2 + X + 2 +np.random.randn(m,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
# degree: 몇 승(차)항까지 추가할 것인가
poly_features= PolynomialFeatures(degree=2, include_bias=False)
X_poly= poly_features.fit_transform(X)

#X_poly: X와 X의 제곱항을 특성으로 추가
print(X[0], X_poly[0])
```

```python
from sklearn.linear_model import LinearRegression 
lin_reg= LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_

#실제: 0.5X^2+ X + 2 + 가우시안 잡음
#모델: coef_[0]X^2+ coef_[1]X +intercept_
```

# 일반화 성능 확인하기
### 교차검증
- 훈련데이터에서는 성능이 좋고, 교차 검증 점수가 나쁘다면 과대적합
- 양쪽에서 모두 나쁘다면 과소적합

### 학습곡선
- 훈련세트와 검증세트의 모델성능 함수 비교

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

def plot_learning_curve(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors= [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict= model.predict(X_train[:m])
        y_val_predict= model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors),"r-+", linewidth=2, label="훈련 세트")
    plt.plot(np.sqrt(val_errors),"b-", linewidth=4, label="검증 세트")
```

```python
lin_reg= LinearRegression()
plot_learning_curve(lin_reg, X, y)

# 과소 적합 그래프 특징
# 검증 성능이 더 좋고, 데이터가 추가되어도 두 그래프 차이가 줄어들지않는다.
# 복잡한 모델, 더 나은 특성을 선택해야함
```

```python
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
    ("lin_reg", LinearRegression())
])

plot_learning_curve(polynomial_regression, X, y)
# 과대적합 특징
# 훈련데이터의 오차가 선형 회귀모델보다 훨씬 낮고, 검증 오차보다도 훨씬 낮음
```

# 규제적용
- 일반화 성능 향상 기법
- 방법1: 가중치 규제항 추가
- 방법2: 조기종료


### 릿지 회귀
- 선형회귀 비용함수에 alpha * Norm2 규제항 추가
    - J 야코비안 행렬: 각 가중치에 대한 도함수 행렬
    - J(x)= MSE(x)+a * sum(|x|^2)
- 규제항은 훈련중에만, 편향 제외한 가중치에만 적용됨
- 이 릿지회귀 또한 정규방정식 또는 경사하강법으로 가중치 행렬 구할 수 있음
    - 행렬분해 - cholesky
- sklearn.linear_model의 Ridge
    - penalty에 사용할 규제 지정

```python
from sklearn.linear_model import Ridge

ridge_reg= Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]])
```

```python
from sklearn.linear_model import SGDRegressor

sed_reg= SGDRegressor(penalty='l2')
sed_reg.fit(X, y)
sed_reg.predict([[1.5]])
```

### 라쏘 회귀
- l1 norm 사용한 규제
    - 덜 중요한 특성의 가중치가 0이 되는 경향이 있음(희소행렬)
    - alpha값이 릿지에 비해 작다.
- 경사하강법 적용시 안정적으로 수렴하는(최적점에 가까워질수록 그래디언트가 작아지는)릿지와 달리 최적점 근처에서 요동친다.
    - 서브그레디언트 벡터 도입: 절댓값 함수는 미분 불가능한 원점 발생, sin으로 대체

```python
from sklearn.linear_model import Lasso

lasso_reg= Lasso(alpha=0.1)
# 동일 sed_reg= SGDRegressor(penalty='l1')
lasso_reg.fit(X,y)
lasso_reg.predict([[1.5]])
```

### 엘라스틱넷
- 릿지와 라쏘 혼합비율 r에 따라 절충해 사용
- 보통 릿지를 사용 
- 특성 수가 훈련 때보다 많거나 특성 몇개가 강한 연관이 있는 경우 라쏘 문제가 되므로 보통 엘라스틱넷 선호

```python
from sklearn.linear_model import ElasticNet

elastic_net= ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, y)
elastic_net.predict([[1.5]])
```

### 조기종료
- 반복학습 알골리즘 과대적합지점(검증에러 최소지점)에서 훈련 중지함으로써 규제
- 성능 곡선이 매끄럽지않아 파악이 어려운 경우
    - 검증에러가 일정시간 최솟값보다 클 때(더 나아지지 않을 때) 정지 후 최소 모델로 복원

```python
from copy import deepcopy
from sklearn.preprocessing import StandardScaler

#데이터 준비
poly_scaler = Pipeline([
    ('poly_features', PolynomialFeatures(degree=90, include_bias=False)),
    ('std_scaler', StandardScaler())
])
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
X_train_poly_scaled= poly_scaler.fit_transform(X_train)
X_val_poly_scaled= poly_scaler.transform(X_val)

# warm_start=True: fit 호출시 처음부터 다시 시작하지 않고 이전 모델 파라미터에서 이어서 함
# tol: None이 아닌 경우 연속 epoch 동안 (loss > best_loss - tol)일 때 교육이 중지
sgd_reg=SGDRegressor(max_iter=1, tol=1e-3, warm_start=True, penalty=None, learning_rate='constant', eta0=0.0005)

minimum_val_error= float("inf")
best_epoch= None
best_model= None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error= mean_squared_error(y_val, y_val_predict)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch= epoch
        best_model= deepcopy(sgd_reg)
```
